```python
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
```

```python
l1=os.listdir("/content/images")  # path to directory where all train images are held
```

```python
l2=os.listdir("/content/test/6000 images for testing")  # path to directory where all test images are held
```

```python
print(len(l1))
print(len(l2))
```

Mixing the two sets of images I had (60K for training and 6K for images to generate randomness)

```python
df1=pd.read_excel("/content/FemalePCs.xlsx") # label file1
df2=pd.read_excel("/content/MalePCs.xlsx") # label file2
```

```python
# creating a list all names in our dataframe
nameM=[]
nameF=[]

for i in range(len(df1)):
  nameF.append(df1.Filename.iloc[i][:10])

for i in range(len(df2)):
  nameM.append(df2.Filename.iloc[i][:10])
```

```python
# creating dictionary train
d1={}
d2={}
for i in tqdm(range(len(l1))):
  if l1[i][:10] in nameF:
    d1[l1[i]]=np.asarray(df1.iloc[nameF.index(l1[i][:10])][1:11])
  else:
    d2[l1[i]]=np.asarray(df2.iloc[nameM.index(l1[i][:10])][1:11])
```

```python
# creating dictionary test
dt1={}
dt2={}
for i in tqdm(range(len(l2))):
  if l2[i][:10] in nameF:
    dt1[l2[i]]=np.asarray(df1.iloc[nameF.index(l2[i][:10])][1:11])
  else:
    dt2[l2[i]]=np.asarray(df2.iloc[nameM.index(l2[i][:10])][1:11])
```

```python
preprocessedF=pd.DataFrame.from_dict(d1,orient='index')
preprocessedM=pd.DataFrame.from_dict(d2,orient='index')
```

```python
preprocessedtestF=pd.DataFrame.from_dict(dt1,orient='index')
preprocessedtestM=pd.DataFrame.from_dict(dt2,orient='index')
```

```python
print(len(preprocessedF))
print(len(preprocessedM))
print(len(preprocessedtestF))
print(len(preprocessedtestM))
```

```python
preprocessedF["name"]=preprocessedF.index
preprocessedM["name"]=preprocessedM.index
preprocessedtestF["name"]=preprocessedtestF.index
preprocessedtestM["name"]=preprocessedtestM.index
```

```python
preprocessedtestM.head()
```

```python
preprocessedF.reset_index(inplace=True)
preprocessedM.reset_index(inplace=True)
preprocessedtestF.reset_index(inplace=True)
preprocessedtestM.reset_index(inplace=True)
```

```python
preprocessedF.drop(columns=["index"],inplace=True)
preprocessedM.drop(columns=["index"],inplace=True)
preprocessedtestF.drop(columns=["index"],inplace=True)
preprocessedtestM.drop(columns=["index"],inplace=True)
```

```python
preprocessedF.to_csv('all_labels_female_corrected.csv',index=False)
preprocessedM.to_csv('all_labels_male_corrected.csv',index=False)
preprocessedtestF.to_csv('all_labels_female_test.csv',index=False)
preprocessedtestM.to_csv('all_labels_male_test.csv',index=False)
```
